## 1. `while è¿´åœˆ`:

### èªæ³•çµæ§‹:

`while` æ¢ä»¶å¼:  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;`åŸ·è¡Œçš„ç¨‹å¼å€å¡Š` (éœ€ç¸®æ’)

```python
#ç¯„ä¾‹: å€’æ•¸è¨ˆæ™‚å™¨
n = 5
while n > 0:
    print(n)
    n = n - 1
print('Blastoff!')
```

ç•¶æ¢ä»¶æ˜¯`True`æ™‚ï¼Œå°±æœƒä¸æ–·åŸ·è¡Œï¼Œç›´åˆ°æŸå€‹æ™‚æ©Ÿä¸»å‹•ä½¿ç”¨`break`è·³å‡ºè¿´åœˆã€‚

5. å¦‚æœè¼¸å…¥çš„ä¸æ˜¯`done`ï¼Œå°±å°å‡ºä½¿ç”¨è€…è¼¸å…¥çš„å…§å®¹`line`ï¼›

## 2. `for è¿´åœˆ`:
### èªæ³•çµæ§‹:

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; ç¨‹å¼å€å¡Š (éœ€è¦ç¸®æ’)

### é€²ä¸€æ­¥èªªæ˜:
`for` è®Šæ•¸ `in` åºåˆ—:  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; ç¨‹å¼å€å¡Š (éœ€è¦ç¸®æ’)  

+ `è®Šæ•¸(variable)`: æ¯æ¬¡è¿´åœˆä¸­ä»£è¡¨ç›®å‰å…ƒç´ çš„è®Šæ•¸ï¼›

### è£œå……èªªæ˜:
åœ¨é€™è£¡ï¼Œ`åºåˆ—(sequence)`æ˜¯æŒ‡**å¯ä»¥ä¸€å€‹ä¸€å€‹å–å‡ºè£¡é¢å…ƒç´ **çš„è³‡æ–™å‹æ…‹ï¼Œå¸¸è¦‹åŒ…æ‹¬ï¼š  
+ `list`: [ 3, 5, 7 ] â†’ `(æœ€å¸¸è¦‹çš„åºåˆ—é¡å‹)`

+ `range()`: range(5)ï¼ŒæŒ‡æ•¸å­—0~4 â†’ `å¸¸æ­é…forä½¿ç”¨`

```python
# ç¯„ä¾‹ (æ‰¾å‡ºæœ€å¤§å€¼)

largest_so_far = -1
print('Before', largest_so_far)

for the_num in [9, 41, 12, 3, 74, 15]:
    if the_num > largest_so_far:
        largest_so_far = the_num
    print(largest_so_far, the_num)

print('After', largest_so_far)
```

`print('After', largest_so_far)` : å°å‡ºæœ€çµ‚çš„æœ€å¤§å€¼çµæœ  

**ç¯„ä¾‹å°çµ:**

## 3. `breakèªå¥`:

### ğŸ” èˆ‡`while`çš„æ­é…:
+ é€šå¸¸ç”¨åœ¨`while True:`é€™é¡ç„¡é™è¿´åœˆä¸­ï¼Œæ­é…`break`ä½œç‚ºçµæŸæ¢ä»¶ã€‚

```python
# ç¯„ä¾‹

while True:
    line = input('> ')
    if line == 'done':
        break
    print(line)
print('Done!')
```
